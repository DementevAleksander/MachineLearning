### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
arr = np.arange(10, 50)
arr_reversed = np.flip(arr)
even_elements = arr_reversed[arr_reversed % 2 == 0]
even_elements
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
arr1 = np.random.rand(6, 4)
arr2 = np.random.rand(4, 3)
result = np.dot(arr1, arr2)
result
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
# Создаем случайный двумерный массив 7x7
arr = np.random.rand(7, 7)

# Находим среднее значение
mean_val = np.mean(arr)

# Находим стандартное отклонение
std_dev = np.std(arr)

print("Среднее значение:", mean_val)
print("Стандартное отклонение:", std_dev)

# Нормализуем массив
normalized_arr = (arr - mean_val) / std_dev
normalized_arr
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
tips = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
print(tips.head(5))
```

### Узнайте сколько всего строчек и колонок в данных

```python
# Определение количества строк и столбцов
num_rows = tips.shape[0]
num_cols = tips.shape[1]

print("Количество строк:", num_rows)
print("Количество столбцов:", num_cols)
```

### Проверьте есть ли пропуски в данных

```python
missing_values = tips.isnull().sum()

print("Количество пропущенных значений в каждом столбце:")
print(missing_values)
```

### Посмотрите на распределение числовых признаков

```python
# Описание числовых признаков.

explanations_numerical = '''
Используем метод describe() на числовых признаках, отобранных с помощью метода select_dtypes()
и фильтрации по типам данных float64 и int64.
Метод describe() выводит основные статистические показатели
(количество, среднее, стандартное отклонение, минимальное и максимальное значения, медиану и квартили)
для каждого числового признака.

'''

numeric_features = tips.select_dtypes(include=['float64', 'int64'])
print(explanations_numerical)
print(numeric_features.describe())

# Визуализация распределения числовых признаков

explanations_visualization = '''
Используем функцию boxplot() из библиотеки seaborn для визуализации распределения числовых признаков в виде боксплотов.
Это позволяет увидеть медиану, квартили, а также выбросы и потенциальные аномалии данных.

'''

print(explanations_visualization)
sns.boxplot(data=numeric_features)
```

### Найдите максимальное значение 'total_bill'

```python
max_total_bill = tips['total_bill'].max()

print("Максимальное значение 'total_bill':", max_total_bill)
```

### Найдите количество курящих людей

```python
count_smokers = tips['smoker'].value_counts()['Yes']

print("Количество курящих людей:", count_smokers)
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
average_total_bill_by_day = tips.groupby('day')['total_bill'].mean()

print("Средний 'total_bill' по дням:")
print(average_total_bill_by_day)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
# Выбор строк с 'total_bill' больше медианы
filtered_tips = tips[tips['total_bill'] > tips['total_bill'].median()]

# Вычисление среднего 'tip' в зависимости от значения в столбце 'sex'
average_tip_by_sex = filtered_tips.groupby('sex')['tip'].mean()

print("Средний 'tip' по полу:")
print(average_tip_by_sex)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
# Создание словаря для преобразования значений 'smoker' в бинарный формат
smoker_map = {'Yes': 1, 'No': 0}

# Преобразование признака 'smoker' в бинарный формат
tips['smoker_binary'] = tips['smoker'].map(smoker_map)

# Вывод преобразованных данных
print(tips[['smoker', 'smoker_binary']].head())
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
# Построение гистограммы распределения признака 'total_bill'
sns.histplot(tips['total_bill'], kde=True)

# Настройка осей и заголовка гистограммы
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill')

# Отображение гистограммы
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
# Построение scatterplot
sns.scatterplot(data=tips, x='total_bill', y='tip')

# Настройка осей и заголовка scatterplot
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs. Tip')

# Отображение scatterplot
plt.show()
```

### Постройте pairplot

```python
# Построение pairplot
sns.pairplot(tips)

# Отображение pairplot
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
# Построение boxplot
sns.boxplot(data=tips, x='day', y='total_bill')

# Настройка осей и заголовка
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')

# Отображение графика
plt.show()
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
# Разделение данных по категориям 'time'
lunch_tips = tips[tips['time'] == 'Lunch']
dinner_tips = tips[tips['time'] == 'Dinner']

# Построение гистограммы распределения признака 'tip' для обеда (Lunch)
sns.histplot(data=lunch_tips, x='tip', kde=True, color='blue', label='Lunch')

# Построение гистограммы распределения признака 'tip' для ужина (Dinner)
sns.histplot(data=dinner_tips, x='tip', kde=True, color='orange', label='Dinner')

# Настройка осей и заголовка
plt.xlabel('Tip Amount')
plt.ylabel('Count')
plt.title('Distribution of Tips by Time')
plt.legend()

# Отображение графиков
plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# Построение scatterplot для Male
male_tips = tips[tips['sex'] == 'Male']
sns.scatterplot(data=male_tips, x='total_bill', y='tip', hue='smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title('Scatterplot for Male')

# Отображение графика
plt.show()

# Построение scatterplot для Female
female_tips = tips[tips['sex'] == 'Female']
sns.scatterplot(data=female_tips, x='total_bill', y='tip', hue='smoker')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.title('Scatterplot for Female')

# Отображение графика
plt.show()
```
